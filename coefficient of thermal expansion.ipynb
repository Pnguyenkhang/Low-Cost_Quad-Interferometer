```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
data = pd.read_csv('/Users/smallwoodlab/Desktop/interferometer_animation/trial_26-05:10:22.csv')

x = data['time']
y1 = data['PD1']
y2 = data['PD2']
y3 = data['theta']
y4 = data['TEMP']
y5 = data['expansion']
```

```python
print(len(x))
start = 0
stop = len(x)-50000
#plt.plot(y1[start:stop],y2[start:stop])
plt.title('PD1 v PD2')
plt.xlabel('Photodetector 1 (V)')
plt.ylabel('Photodetector 2 (V)')
plt.plot(y5[start:stop])
```

```python
def averagingFunc(temp,n):
    new_aves = list()
    n_div = n # divisions
    curr_step = 0
    while curr_step < len(temp):
        val = np.average(temp[curr_step:curr_step+n_div])
        curr_step = curr_step + n_div
        new_aves.append(val)
    return new_aves

ave_temp = averagingFunc(y4[start:stop],3)
ave_expansion = averagingFunc(y5[start:stop],5)
ave_temp = np.array(ave_temp)
```

```python
plt.plot(y4[start:stop]*100,y5[start:stop])
#plt.plot(ave_temp*100,ave_expansion)
plt.suptitle(('Coefficient of Linear Thermal Expansion: '+ str(round(np.polyfit(y4[start:stop]*100,y5[start:stop],1)[0]*(1/0.24),7))), fontsize=12, fontweight='bold')
#plt.suptitle(('Coefficient of Linear Thermal Expansion: '+ str(round(np.polyfit(ave_temp*100,ave_expansion,1)[0]*(1/0.24),7))), fontsize=12, fontweight='bold')
plt.title('Temperature v Expansion')
plt.xlabel('Temperature (C)')
plt.ylabel('Expansion (m)')
```
